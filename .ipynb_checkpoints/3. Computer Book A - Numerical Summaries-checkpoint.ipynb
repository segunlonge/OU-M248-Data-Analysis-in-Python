## Computer activity 14: Weight change in clinical trials

Obtain a horizontal boxplot of weight change for all the participants, together with the following numerical summaries of the data: number of values, 
number of missing values, sample mean, sample median, sample standard deviation and sample interquartile range.

*Datasource: Worksheet with id, control group, gender and weight change delta columns*
<br>
*Note: Group 0 = Control and Group 1 = Treatment*

```python
import pandas as pd

# Load data from csv
df = pd.read_csv(r'C:\Python\OU M248 Computer Activities using Python\Data files\response-inhibition.csv')

# Convert values in the Weight change column to numeric; convert non numeric values on error to NaN
# Use coerce to set invalid parsing to NaN
df["Weight change"] = pd.to_numeric(df["Weight change"], errors='coerce')

# use round function() to round to 2 decimal places
mean_weight_change = round(df['Weight change'].mean(),3)
std_weight_change = round(df['Weight change'].std(),3)
median_weight_change = round(df['Weight change'].median(),3)
iqr_weight_change = round(df['Weight change'].quantile(0.75) - df['Weight change'].quantile(0.25),3)
missing_count = round(df['Weight change'].isna().sum(),3)
non_missing_count = round(df['Weight change'].notna().sum(),3)

print("Summary statistics for weight change trial")

print(f"Mean (Weight Change): {mean_weight_change}")
print(f"Standard deviation (Weight Change): {std_weight_change}")
print(f"Median (Weight Change): {median_weight_change}")
print(f"Interquartile range (Weight Change) : {iqr_weight_change}")
print(f"Missing count (Weight Change): {missing_count}")
print(f"Standard deviation (Weight Change): {non_missing_count}")
```

#### Commentary

Weight change data is not available for 2 of the trial participants. Note that the median and the mean are roughly similar which means that the data is symmetric but not necessarily normaly distributed.

The symmetry is indicated in the box plot below.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from csv
df = pd.read_csv(r'C:\Python\OU M248 Computer Activities using Python\Data files\response-inhibition.csv')

# Return the first 5 rows of the dataframe
print(df.head())

# Convert 'Weight change' column to numeric, coercing errors to NaN
df['Weight change'] = pd.to_numeric(df['Weight change'], errors='coerce')

# Drop rows with NaN values in 'Weight change' column
# df = df.dropna(subset=['Weight change'])

# Create the boxplot
df.boxplot(column='Weight change', vert=False)
# df.boxplot(column='Weight change', by='Group', vert=False)

plt.xlabel('Weight change')
plt.show()
```

#### Commentary 

The data appears to be symmetric as the whiskers are broadly the same length and the boxes are roughly similar on each side. There are some outliers in the dataset with one having a large weight gain and three having a large weight loss.

Given the outliers it would be better to rely on the mean and the interquartile range to get a better summary of the data. The standard deviation is quite high given the outliers but going by the mean it can be said that the average weight change is -0.244 kg against the interquartile range of 1.650 kg meaning the typical participant lost weight lost some weight in during the two week trial.

## Computer activity 15: More on obtaining mumerical summaries

Obtain the following descriptive statistics for each group in the clinical trial treament program: <br> Mean, Standard deviation, Median, Interquartile range, N nonmissing, N missing.

```python
import pandas as pd

# Load data from csv
df = pd.read_csv(r'C:\Python\OU M248 Computer Activities using Python\Data files\response-inhibition.csv')

# Convert values in the Weight change column to numeric; convert non numeric values on error to NaN
# Use coerce to set invalid parsing to NaN
df["Weight change"] = pd.to_numeric(df["Weight change"], errors='coerce')

# Function to calculate the Interquatile Range
def iqr(series):
    return series.quantile(0.75) - series.quantile(0.25)

summary = df.groupby("Group")["Weight change"].agg(
    mean="mean",
    std="std",
    median="median",
    IQR=iqr,
    # takes a parameter x i.e. the Pandas Series from each group and applies isna() and sum() to get sum of missing
    # values per group
    missing=lambda x: x.isna().sum()
)

print(summary)

```
