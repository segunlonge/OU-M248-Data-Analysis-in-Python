## Computer activity 4: Quality of tattoo removal

In this activity you will obtain a bar chart similar to the one in Figure 1 of Unit 1 for the data on the quality of tattoo removal.

```python
import pandas as pd
import matplotlib.pyplot as plt

#load data from csv
df = pd.read_csv(r'C:\Python\OU M248 Computer Activities using Python\Data files\tattoos.csv')
#print(df.head()) #return the first 5 rows of the dataframe

#Sort the count in ascending numerical order
counts = df['Score'].value_counts().sort_index()

#Plot the counts as a bar chart
counts.plot(kind='bar')
#rotate the x-axis upright
plt.xticks(rotation = 360)

#Plot the labels and title
plt.xlabel('Score')
plt.ylabel('Count')
plt.title('Chart of Score')
plt.tight_layout()
#plt.show
```

## Computer activity 6: Total UK workforce

In this activity, you will obtain a bar chart for the data on the total UK workforce similar to the one in Figure 3 of Unit 1 (*the activity asks for the plot of the total for each work category*)

```python
import pandas as pd
import matplotlib.pyplot as plt

#load data from csv
workforce = pd.read_csv(r'C:\Python\OU M248 Computer Activities using Python\Data files\workforce.csv')
#print(df.head()) #return the first 5 rows of the dataframe

# settings for printing the entire contents of dataframe
pd.set_option('display.max_rows', None) # show all rows
pd.set_option('display.max_columns', None) # show all columns

df = pd.DataFrame(workforce)
# total_sorted = df.sort_values(by="Total", ascending=False)
# total_sorted = df.sort_values("Total")

# Sort the data by the 3rd column named "Total"
df_sorted = df.copy()
column_name = df_sorted.columns[3]
df_sorted = df_sorted.sort_values(by=column_name, ascending=False)

print(df_sorted)

# set the columns to use for the plot
X = list(df_sorted.iloc[:, 0])
Y = list(df_sorted.iloc[:, 3])

# plot the bar chart using the bar() method
plt.bar(X, Y, color='g')
plt.title("Total UK Workforce")
plt.xlabel("Occupation Type")
plt.ylabel("Frequency (millions)")
plt.xticks(rotation=45, ha='right')  # rotate the xlabels to be readable

# show the plot
plt.show()
```
